```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

table = soup.find("table", {"class": "wikitable sortable"})

rows = table.find_all("tr")

for row in rows[1:]:
    cells = row.find_all("td")

    rank = cells[0].text.strip()
    name = cells[1].text.strip()
    artist = cells[2].text.strip()
    upload_date = cells[3].text.strip()
    views = cells[4].text.strip()

    print("Rank:", rank)
    print("Name:", name)
    print("Artist:", artist)
    print("Upload date:", upload_date)
    print("Views:", views)
    print("-----------------------------------")
```

```python
url = "https://www.bcci.tv/international/fixtures"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

fixtures_container = soup.find("div", class_="js-list")

if fixtures_container:

    fixtures = fixtures_container.find_all("div", class_="fixture")

    for fixture in fixtures:
        title = fixture.find("p", class_="fixture__additional-info").text.strip()
        series = fixture.find("span", class_="u-unskewed-text").text.strip()
        place = fixture.find("p", class_="fixture__additional-info").find_next("span").text.strip()
        date = fixture.find("div", class_="fixture__full-date").text.strip()
        time = fixture.find("div", class_="fixture__time").text.strip()

        print("Match Title:", title)
        print("Series:", series)
        print("Place:", place)
        print("Date:", date)
        print("Time:", time)
        print("-----------------------------------")
else:
    print("Fixtures not found on the BCCI website.")
```

```python

url = "https://www.bcci.tv/resources/data/men/matches/fixturesFull?comptype=2"
response = requests.get(url)

if response.status_code == 200:
    fixtures_data = response.json()

    for fixture in fixtures_data:
        title = fixture["short_name"]
        series = fixture["series_name"]
        place = fixture["venue_name"]
        date = fixture["formatted_date"]
        time = fixture["formatted_time"]

        print("Match Title:", title)
        print("Series:", series)
        print("Place:", place)
        print("Date:", date)
        print("Time:", time)
        print("-----------------------------------")
else:
    print("Failed to retrieve fixtures data from the BCCI website.")
```

```python

url = "http://statisticstimes.com/economy/india-statistics.php"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

gdp_table = soup.find("table", id="table_id")

if gdp_table:
    rows = gdp_table.find_all("tr")

    for row in rows[1:]:
        columns = row.find_all("td")
        rank = columns[0].text.strip()
        state = columns[1].text.strip()
        gdp_18_19 = columns[2].text.strip()
        gdp_19_20 = columns[3].text.strip()
        share_18_19 = columns[4].text.strip()
        gdp_billion = columns[5].text.strip()

        print("Rank:", rank)
        print("State:", state)
        print("GSDP (18-19) - at current prices:", gdp_18_19)
        print("GSDP (19-20) - at current prices:", gdp_19_20)
        print("Share (18-19):", share_18_19)
        print("GDP ($ billion):", gdp_billion)
        print("-----------------------------------")
else:
    print("State-wise GDP table not found on the website.")
```

```python
url = "https://api.github.com/search/repositories"
params = {
    "q": "stars:>500",  
    "sort": "stars",
    "order": "desc",
}
response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()

    for repository in data["items"]:
        title = repository["name"]
        description = repository["description"]
        contributors_count = repository["contributors_url"]
        language = repository["language"]

        print("Repository Title:", title)
        print("Description:", description)
        print("Contributors Count:", contributors_count)
        print("Language Used:", language)
        print("-----------------------------------")
else:
    print("Failed to retrieve trending repositories from GitHub.")
```

```python
url = "https://www.billboard.com/charts/hot-100"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

chart_container = soup.find("div", class_="chart-list container")

chart_rows = chart_container.find_all("div", class_="chart-list-item")

for row in chart_rows:
    song_name = row.find("span", class_="chart-list-item__title-text").text.strip()
    artist_name = row.find("span", class_="chart-list-item__artist").text.strip()

    last_week_rank = row.find("div", class_="chart-list-item__last-week").text.strip()
    peak_rank = row.find("div", class_="chart-list-item__weeks-on-chart").text.strip()
    weeks_on_board = row.find("div", class_="chart-list-item__weeks-at-one").text.strip()

    print("Song Name:", song_name)
    print("Artist Name:", artist_name)
    print("Last Week Rank:", last_week_rank)
    print("Peak Rank:", peak_rank)
    print("Weeks on Board:", weeks_on_board)
    print("-----------------------------------")
```

```python
url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

table = soup.find("table")

rows = table.find_all("tr")

for row in rows[1:]:
    columns = row.find_all("td")

    if len(columns) >= 5:
        book_name = columns[0].text.strip()
        author_name = columns[1].text.strip()
        volumes_sold = columns[2].text.strip()
        publisher = columns[3].text.strip()
        genre = columns[4].text.strip()
 
        print("Book Name:", book_name)
        print("Author Name:", author_name)
        print("Volumes Sold:", volumes_sold)
        print("Publisher:", publisher)
        print("Genre:", genre)
        print("-----------------------------------")
```

```python
url = "https://www.imdb.com/list/ls095964455/"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

series_list = soup.find_all("div", class_="lister-item-content")

for series in series_list:
    name = series.find("h3", class_="lister-item-header").a.text
    
    year_span = series.find("span", class_="lister-item-year").text.strip("()")
    
    genre = series.find("span", class_="genre").text.strip()
    
    run_time = series.find("span", class_="runtime").text.strip()
    
    ratings = series.find("span", class_="ipl-rating-star__rating").text
    
    votes = series.find("span", attrs={"name": "nv"}).text
    
    print("Name:", name)
    print("Year Span:", year_span)
    print("Genre:", genre)
    print("Run Time:", run_time)
    print("Ratings:", ratings)
    print("Votes:", votes)
    print("-----------------------------------")
```

```python
url = "https://archive.ics.uci.edu/"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

link = soup.find("a", text="Show All Dataset")["href"]

show_all_url = url + link

show_all_response = requests.get(show_all_url)

show_all_soup = BeautifulSoup(show_all_response.content, "html.parser")

rows = show_all_soup.find_all("tr", valign="top")

for row in rows:
    columns = row.find_all("td")
    
    dataset_name = columns[0].text.strip()
    data_type = columns[1].text.strip()
    task = columns[2].text.strip()
    attribute_type = columns[3].text.strip()
    instances = columns[4].text.strip()
    attributes = columns[5].text.strip()
    year = columns[6].text.strip()

    print("Dataset Name:", dataset_name)
    print("Data Type:", data_type)
    print("Task:", task)
    print("Attribute Type:", attribute_type)
    print("Number of Instances:", instances)
    print("Number of Attributes:", attributes)
    print("Year:", year)
    print("-----------------------------------")
```

```python
url = "https://www.naukri.com/hr-recruiters-consultants"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

rows = soup.find_all("div", class_="recSec")

for row in rows:
    name = row.find("span", class_="fl").text.strip()
    designation = row.find("span", class_="designation").text.strip()
    company = row.find("p", class_="highlightable").text.strip()
    skills = row.find("div", class_="hireSec highlightable").text.strip()
    location = row.find("small", class_="ellipsis").text.strip()

    print("Name:", name)
    print("Designation:", designation)
    print("Company:", company)
    print("Skills they hire for:", skills)
    print("Location:", location)
    print("-----------------------------------")
```
